# Groupby

The groupby method allows you to group rows of data together and call aggregate functions

```python
import pandas as pd
```

```python
df = pd.read_csv('data/titanic.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Pclass'].value_counts()
```

```python
df.groupby('Pclass')['Pclass'].count()
```

```python
# number of survived people for each pclass
df.groupby('Pclass')['Survived'].sum()
```

```python
# number of survived people and fare for each pclass
df.groupby('Pclass')[['Survived', 'Fare']].sum()
```

```python
# group by multiple columns
# group by pclass and sex

df.groupby(['Pclass', 'Sex'])['Survived'].sum()
```

```python
df.groupby(['Pclass', 'Sex'], as_index = False)['Survived'].sum()
```

```python
df.groupby(['Pclass'])['Survived'].describe()
```

```python
## multiple function

df.groupby(['Pclass'])['Survived'].agg(['mean', 'sum', 'std', 'median', 'min', 'max'])
```

```python
## multiple function

df.groupby(['Pclass'])[['Survived', 'Fare']].agg(['mean', 'sum', 'min', 'max'])
```

```python
df.groupby(['Pclass'], as_index = False)['Survived'].mean()
```

```python
df.groupby(['Pclass'], as_index = False)['Fare'].sum()
```

```python

df.groupby(['Pclass']).agg({'Survived' : 'mean', 'Fare' : 'sum'})
```
